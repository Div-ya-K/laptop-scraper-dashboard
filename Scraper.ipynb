```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://webscraper.io/test-sites/e-commerce/static/computers/laptops")
```

```python
for i in range(1,21):
    a=requests.get(f"https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={i}")
    with open(f"htmls/page{i}.html","w",encoding="utf-8") as f:
        f.write(a.text)
        print(f"Downloaded page {i} sucessfully")
```

```python
items = []
for file_name in os.listdir("htmls"):
    if file_name.endswith(".html"):
        file_path = os.path.join("htmls", file_name)
        
        with open(file_path, 'r', encoding='utf-8') as f:
            html = f.read()
            soup = BeautifulSoup(html, 'html.parser')
            
            products = soup.select("div.card.thumbnail")
            
            for product in products:
                title_tag = product.find("a", class_="title")
                title = title_tag.text.strip() if title_tag else "No title"

                desc_tag = product.find("p", class_="description")
                description = desc_tag.text.strip() if desc_tag else "No description"

                price_tag = product.find("h4", class_="price")
                price = price_tag.text.strip().replace("$", "") if price_tag else "0"

                rating = len(product.select("div.ratings span.ws-icon-star"))

                items.append({
                    "Title": title,
                    "Price": float(price),
                    "Description": description,
                    "Rating": rating,
                
                })
```

```python
import pandas as pd
df = pd.DataFrame(items)
```
